```python

import pandas as pd
import re

# Read the text file
with open("training_stats_val.txt", 'r') as file:
    lines = file.readlines()

# Initialize empty lists to store data
epochs = []
loss = []
val_loss = []
auc_1 = []
val_auc_1 = []

# Regular expressions to match and extract the values
epoch_pattern = r'Epoch (\d+)/\d+'
metrics_pattern = r'loss: ([\d.]+) - auc_1: ([\d.]+) - val_loss: ([\d.]+) - val_auc_1: ([\d.]+)'

for line in lines:
    epoch_match = re.search(epoch_pattern, line)
    if epoch_match:
        epochs.append(int(epoch_match.group(1)))
    
    metrics_match = re.search(metrics_pattern, line)
    if metrics_match:
        loss.append(float(metrics_match.group(1)))
        auc_1.append(float(metrics_match.group(2)))
        val_loss.append(float(metrics_match.group(3)))
        val_auc_1.append(float(metrics_match.group(4)))

# Create a pandas DataFrame
data = {
    'Epoch': epochs,
    'loss': loss,
    'val_loss': val_loss,
    'auc_1': auc_1,
    'val_auc_1': val_auc_1
}

df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('epoch_data.csv', index=False)
```
